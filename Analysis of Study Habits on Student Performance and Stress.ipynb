```r
total_hours <- tapply(data$Hours_Studying_Per_Day, data$Gender, sum)
hours_studying_by_gender <- data.frame(Gender = names(total_hours),
                                       Hours_Studying_Per_Day = as.numeric(total_hours))

hours_studying_by_gender$Percentage <- (hours_studying_by_gender$Hours_Studying_Per_Day / sum(hours_studying_by_gender$Hours_Studying_Per_Day)) * 100

ggplot(hours_studying_by_gender, aes(x = "", y = Hours_Studying_Per_Day, fill = Gender)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar("y", start = 0) +
  labs(title = "Distribution of Hours Studying per Day by Gender",
       fill = "Gender",
       y = "Total Hours Studying per Day") +
  geom_text(aes(label = paste0(round(Percentage), "%")), position = position_stack(vjust = 0.5)) +
  theme_void()
```
